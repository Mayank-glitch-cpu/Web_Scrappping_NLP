```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
from bs4.element import Comment

input_data= pd.read_excel('/content/Input.xlsx')
#print(input_data)
Url_id= input_data.iloc[:,0]
Url= input_data.iloc[:,1]

data_extracted={}

for i in range(0,len(Url_id)):
    key= Url_id[i]
    value= Url[i]
    data_extracted[key]=value

#print(data_extracted[Url_id[0]])
text_extracted={}
for i in range(0,len(Url_id)):
     #print(Url_id)
     #print(Url)
     
     key2=Url_id[i]
     # URL to be scraped
     url = data_extracted[Url_id[i]]
     
     # Make a request to the URL
     response = requests.get(url)
     
     # Parse the HTML content using BeautifulSoup
     soup = BeautifulSoup(response.content, "html.parser")
     
     # Find the main article content
     main_article = soup.find("article", class_="post")
     
     # Extract the text from the main article content
     if main_article:
         # Remove any unnecessary elements from the main article content
         for element in main_article(["script", "style", "footer"]):
             element.decompose()
             
         # Remove the related articles, most popular and recent comments sections
         for element in main_article.find_all("div", class_=["td-related-span4", "td_block_inner td-column-1",
                                                             "td_block_wrap tdb_single_tags tdi_133 td-pb-border-top td_block_template_1",
                                                             "td-post-sharing-visible","tdb-next-post tdb-next-post-bg tdb-post-prev",
                                                             "tdb-next-post tdb-next-post-bg tdb-post-next",
                                                             "td_block_wrap td_flex_block_1 tdi_141 td_with_ajax_pagination td-pb-border-top td_block_template_1 td_flex_block",
                                                             "td_block_wrap td_flex_block_1 tdi_146 td_with_ajax_pagination td-pb-border-top td_block_template_1 td_flex_block",
                                                             "td_block_wrap vc_wp_recentcomments tdi_147 td-pb-border-top td_block_template_1"]):
             element.decompose()
     
         # Extract the relevant text from the main article content
         text_elements = main_article.find_all(string=True)
         relevant_text = [element.strip() for element in text_elements if not isinstance(element, Comment)]
     
         # Join the relevant text into a string
         text = " ".join(relevant_text)
     
         # Split the text into chunks of a fixed length
         chunk_size = 80
         text_chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
     
         # Join the text chunks horizontally
         text_horizontal = "\n".join(text_chunks)
     
         # Print the final text
         #print(text_horizontal)
     else:
         print("Main article content not found for Url Id"+str(Url_id[i]))
     text_extracted[key2]=text_horizontal  

print('*****************Extraction_complete************************')
#for i in range(0,len(Url_id)):
#  print(text_extracted[Url_id[i]])
#  print('**************Text'+str(i)+'******************')   
    
    
```

Filtering Stop Words


```python
filtered_words = []
filter_dict={}
for i in range(0,len(Url_id)):
   key4=Url_id[i]
   for word in token_dict[Url_id[i]]:
          if word.lower() not in stop_words:
              filtered_words.append(word)
   filter_dict[key4]=filtered_words
   filtered_words=[]
   
print(filtered_words)
print(filter_dict[Url_id[0]])
word_count_after_cleaning={}
#n_count={}
for i in range(0,len(Url_id)):
  key8= Url_id[i]
  word_count_after_cleaning[key8]=len(filter_dict[Url_id[i]])
print(word_count_after_cleaning[Url_id[0]]) 
```
